# Practice Peer-graded Assignment: Create a Labelled Plot of the Happiness Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("happyscore_income.csv")

df
```

```python
plt.figure(figsize=(12,12))

sns.scatterplot(x="happyScore", y="GDP", data=df, size="income_inequality")
plt.title("Happy Scores vs GDP")

plt.xlabel("Happiness Score")
plt.ylabel("GDP")

plt.show()
```
